```python
def clean(raw):
    match= re.findall(r'(?:(?<=^)|(?<=[^A-Za-z.,]))[A-Za-z.,]+(?: [A-Za-z.,]+)*(?:(?=[^A-Za-z.,])|(?=$))', raw)
    if len(match)==0:
        return raw
    else:
        return match[0]
```

```python
#cleaning comment
df['comment']=df['comment'].apply(clean)
```

```python
match= re.findall(r'[a-zA-Z]', 'Combination of beast and beauty')
print(match)
match= re.findall(r'[a-zA-Z]', 'ð')
print(match)
```

```python
def clean_emoji(raw):
    match= re.findall(r'[a-zA-Z]', raw)
    if len(match)==0:
        return '---emojis---'
    else:
        return raw
```

```python
#cleaning comment
df['comment']=df['comment'].apply(clean_emoji)
```

```python
def extract(raw):
    names=re.findall(r'[A-Z][a-z]*',raw)
    if len(names[n:])==0:
        return 'Mukul Verma'
    else:
        return " ".join(names[n:])
```

```python
df['reciever']=df['title'].apply(extract)
```

```python
#it will separate replied and commented
def categorise(raw):
    l=re.findall(r'[A-Z]*[a-z]*',raw)
    if l[4]=='replied':
        return 'replied'
    elif l[4]=='commented':
        return 'commented'
```

```python
df['category']=df['title'].apply(categorise)
```

```python
#converting date
df['year']=df['comment_time'].dt.year
df['month']=df['comment_time'].dt.month
df['date']=df['comment_time'].dt.day
df['dayofweek']=pd.Categorical(df['comment_time'].dt.weekday_name,ordered=True,categories=['Sunday','Monday','Tuesday','Wednesday','Thursday','Friday','Saturday'])
df['hour']=df['comment_time'].dt.hour+df['comment_time'].dt.minute/60
df['hour_only']=df['comment_time'].dt.hour
```

```python
def top_commented(n):
    ax=df['reciever'].value_counts().head(n).plot(kind='bar')
    ax.set_xlabel("Number of comments")
    ax.set_ylabel("On whoom post did commented the most")
    ax.set_title("People to which you commented most")
```

```python
def top_commented_others(n):
    bx=df[df['category']=='commented']['reciever'].value_counts().head(n).plot(kind='bar')
    bx.set_xlabel('Commented on the post of these people')
    bx.set_ylabel("Number of comments")
    bx.set_title("Comments on others' posts")
```

```python
def no_commented_particular(x):
    try:
        print("Total number of comments on ",x,"'s posts are:",df['reciever'].value_counts()[x])
    except KeyError:#if person is not in list
        print("No such person/page found")
```

```python
def top_commented_year(n):
    dx=df['year'].value_counts().head(n).plot(kind='bar')
    dx.set_xlabel('Years')
    dx.set_ylabel("Number of comments")
    dx.set_title("Top Years")
```

```python
def top_commented_days_of_week():
    dx=df['dayofweek'].value_counts().plot(kind='bar')
    dx.set_xlabel("Week Days")
    dx.set_ylabel("Number of comments")
    dx.set_title("Days of week you comment the most")
```

```python
def top_commented_dates(n):
    ex=df['date'].value_counts().head(n).plot(kind='bar')
    ex.set_xlabel("Dates")
    ex.set_ylabel("Number of comments")
    ex.set_title("Top dates on which you comment the most")
```

```python
def top_commented_hour(n):
    fx=df['hour_only'].value_counts().head(n).plot(kind='bar')
    fx.set_xlabel("Hours")
    fx.set_ylabel("Number of comments")
    fx.set_title("Top hous in which you comment")
```

```python
def top_commented_date():
    y=df['year'].value_counts().axes[0][0]#top year
    m=df[df['year']==y]['month'].value_counts().axes[0][0]
    print("Most commented date is:",df[df['month']==m]['date'].value_counts().axes[0][0],"-",m,"-",y)
```

```python
def category():
    gx=df['category'].value_counts().plot(kind='bar')
    gx.set_xlabel("Category")
    gx.set_ylabel("Number of comments")
    gx.set_title("Categorise")
```

```python
def time_range(n):
    a=int(input("Enter the starting time:"))
    b=int(input("Enter the ending time:"))
    try:
        ex=df[(df['hour']>=a)&(df['hour']<b)]['reciever'].value_counts().head(n).plot(kind='bar')
        print("Top persons to recive comments between ",a," to ",b," are:")
        ex.set_xlabel("Time(hours)")
        ex.set_ylabel("Number of comments")
        ex.set_title("Time Range")
    except:
        print("No comments in this time period!")
```

```python
def time_range_particular_reciever(x):
    a=int(input("Enter the starting time:"))
    b=int(input("Enter the ending time:"))
    try:
        ex=df[(df['hour']>=a)&(df['hour']<b)]['reciever'].value_counts()[x]
        print("No. of comments recieved by ",x," between ",a," to ",b," are:",ex)
    except:
        print("Either no comments of this person/page in this time period! or No such person exist!")    
```

```python
def all_comments_particular_user(x):
    for i in df[df['reciever']==x]['comment']:
        print(i)
    else:
        print("Person not Exist!")
```

```python
def mention(x):
    for i in df['comment'].values:
        if x in i:
            print(i)
```

```python
print("Choose from:\n1.Top Persons/Pages on whose post you have commented the most(including your replies)\n2.Top Persons/Pages on whose post you have commented the most(excluding your replies)\n3.Number of comments on posts of a particular person\n4.Top years in which you have commented the most\n5.Top days of week on which you have commented the most\n6.Top dates on which you have commented the most\n7.Top hour in which you have commented the most\n8.Overall date on ehich you have commented the most\n9.Categorywise(Commented/Replied)\n10.Comments of number of top recievers in a particyular time range\n11.Comments of a particular reciever ina particular time range\n12.All comments of a particular user\n13.Comments in which a particular user is mentioned\n------------>")
k=int(input())
if k==1:
    n=int(input("Enter the number of top persons/Pages you want to see(including your replies):"))
    top_commented(n)
elif k==2:
    n=int(input("Enter the number of top persons/Pages you want to see(excluding your replies):"))
    top_commented_others(n)
elif k==3:
    #print("Choices are:",df['reciever'].value_counts().axes[0].values)
    x=input("Enter the name of person/page:")
    no_commented_particular(x)
elif k==4:
    n=int(input("Enter the number of years:"))
    top_commented_year(n)
elif k==5:
    top_commented_days_of_week()
elif k==6:
    n=int(input("Enter the number of dates:"))
    top_commented_dates(n)
elif k==7:
    n=int(input("Enter the number of hours:"))
    top_commented_hour(n)
elif k==8:
    top_commented_date()
elif k==9:
    category()
elif k==10:
    n=int(input("Enter the nuber of recievers:"))
    time_range(n)
elif k==11:
    x=input("Enter the name of Person/Page whose comments you want to see for a particular time range:")
    time_range_particular_reciever(x)
elif k==12:
    #print("Choices are:",df['reciever'].value_counts().axes[0].values)
    x=input("Enter the name of Person/Page whose comments you want to see:")
    all_comments_particular_user(x)
elif k==13:
    x=input("Enter the name of Person/Page whose comments you want to see in which he/she/it is mentioned:")
    mention(x)
else:
    print("You entered wrong choice")
```
